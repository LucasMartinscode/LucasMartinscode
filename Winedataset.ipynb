```python
import pandas as pd #Abrindo pandas para tratar base de dados
import random 
import plotly.graph_objects as go

dataset = pd.read_csv('winemag-data-130k-v2.csv')
datasetdropna = dataset.dropna() 
```

```python
datasetdropna.head()
```

Amostragem Aleatória simples

```python
Amostra_aleatoria_simples = datasetdropna.sample(n= 100, random_state=2) # Amostra simples utilizando 100 linhas usando random state para ser variado o número de amostras

Amostra_aleatoria_simples.head()
```

```python
media_price = Amostra_aleatoria_simples['price'].mean() #Descobrindo a média de preços
desvio_padrao_price = Amostra_aleatoria_simples['price'].std() #Descobrindo o desvio padrão de preços
min_price = Amostra_aleatoria_simples['price'].min() #Descobrindo o minimo de preços
max_price = Amostra_aleatoria_simples['price'].max() #Descobrindo o maximo de preços
quartis_price = Amostra_aleatoria_simples['price'].quantile([0.25, 0.5, 0.75]) #Descobrindo a distribuição de preços

# Criando uma tabela com informações de preços

resultados = {
    'Medida': ['Média', 'Desvio Padrão', 'Mínimo', 'Máximo', '25% Quartil', '50% Quartil', '75% Quartil'],
    'Valor': [media_price, desvio_padrao_price, min_price, max_price, quartis_price[0.25], quartis_price[0.5], quartis_price[0.75]]}

Tabela_resultados = pd.DataFrame(resultados)

Tabela_resultados.head()
```
